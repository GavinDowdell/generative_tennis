# setting device
### The torch.device() function and the device string are both used to specify the device on which a PyTorch tensor should be allocated. However, there are some subtle differences between the two.

### The torch.device() function takes a string as its argument, which can be either "cpu" or "cuda". The "cpu" device represents the CPU, while the "cuda" device represents the GPU. The device string, on the other hand, can be either a string or a number. If it is a string, it must be either "cpu" or "cuda". If it is a number, it represents the index of the GPU to be used.

### In general, it is better to use the torch.device() function to specify the device on which a tensor should be allocated. This is because the torch.device() function is more flexible and can be used to specify any device, including custom devices. The device string, on the other hand, is only valid for the CPU and GPU devices.

### Hence can be a string or a function return however either way need to use the .to method for conversion
